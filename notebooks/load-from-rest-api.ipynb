# Loading data in Pandas from the GoldenCheetah REST API

```python
import pandas as pd
from io import StringIO
import requests
import re
```

```python
base_url = 'http://localhost:12021/'
```

```python
athlete_url = base_url + 'david/'
```

## Loading an athlete

```python
req = requests.get(athlete_url)
print("Status code: {}".format(req.status_code))
print("Encoding: {}".format(req.encoding))
```

```python
req.encoding = 'utf-8-sig'
req.encoding
```

```python
ath = pd.read_csv(StringIO(req.text))
ath.columns = [col.strip() for col in ath.columns]  # Remove whitespaces in column names
ath.head()
```

### Select columns

```python
cols = [col.lower() for col in ath.columns.values]
len(cols)
```

```python
ath.columns = cols
```
